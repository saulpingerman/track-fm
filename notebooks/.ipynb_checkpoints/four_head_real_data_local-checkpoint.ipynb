```python
%cd /home/ec2-user
# ─── Cell 1 : imports & utility ────────────────────────────────────────────
import math
from pathlib import Path
import time
import sys

import polars as pl
import torch
from torch import nn, optim
from torch.utils.data import IterableDataset, DataLoader

# make sure your Fourier-head repo is importable
fourier_dir = Path.home() / "repos" / "fourier-head" / "notebooks"
sys.path.insert(0, str(fourier_dir))
from four_head_2D import FourierHead2D      # 2D Fourier head


def latlon_to_local_uv(lat, lon, lat0, lon0, *, half_side_mi: float = 50.0):
    """
    lat, lon, lat0, lon0 : tensors in **degrees** (any shape …)
    Returns
    -------
    uv   : (..., 2) tensor, each component in [-1,1] (clipped)
    logJ : (...)   tensor, log-Jacobian from (u,v) → (lat,lon)
    """
    # convert to radians once
    lat_rad  = torch.deg2rad(lat)
    lat0_rad = torch.deg2rad(lat0)

    R_mi = 69.0  # mean statute-miles per degree latitude
    dx_mi = R_mi * torch.cos(lat0_rad) * (lon - lon0)
    dy_mi = R_mi * (lat - lat0)

    u = dx_mi / half_side_mi
    v = dy_mi / half_side_mi

    # |J| = (∂x/∂u)(∂y/∂v) = (half_side_mi)² / (R*cosφ · R)
    # log|J| is *subtracted* when we move (u,v)→(lat,lon)
    logJ = (
         2.0 * math.log(half_side_mi)
         - 2.0 * math.log(R_mi)
         - torch.log(torch.cos(lat0_rad).clamp_min(1e-6))  # avoid NaN at poles
    )

    uv = torch.stack([torch.clamp(u, -1.0, 1.0), torch.clamp(v, -1.0, 1.0)], dim=-1)
    return uv, logJ
```

```python
# ─── Cell 2 : data-loading helpers ─────────────────────────────────────────
def load_cleaned_data(cleaned_root: str) -> pl.DataFrame:
    p = Path(cleaned_root).resolve()
    files = list(p.rglob("*.parquet"))
    if not files:
        raise FileNotFoundError(f"No parquet files under {p}")
    print(f"🟢 Reading {len(files)} parquet file(s)…", flush=True)
    return pl.read_parquet([str(f) for f in files])


class AISForecastIterableDataset(IterableDataset):
    """
    Streams (past_uv, centre_ll_norm, target_uv, logJ) windows one vessel at a time.
    past_uv          : (seq_len, 2)  in [-1,1]²
    centre_ll_norm   : (2,)          absolute context  (lat/90, lon/180)
    target_uv        : (2,)          in [-1,1]² (clipped if necessary)
    logJ             : scalar        Jacobian term for loss
    """
    def __init__(self, df: pl.DataFrame, *, seq_len: int = 10, half_side_mi: float = 50.0):
        self.df = df.sort(["mmsi", "timestamp"])
        self.mmsis = self.df["mmsi"].unique().to_list()
        self.seq_len = seq_len
        self.half_side_mi = half_side_mi
        print(f"🟢 Dataset ready — {len(self.mmsis)} vessels", flush=True)

    def __iter__(self):
        for m in self.mmsis:
            group = (
                self.df
                .filter(pl.col("mmsi") == m)
                .select(["lat", "lon"])
            )
            coords = torch.tensor(group.to_numpy(), dtype=torch.float32)  # (N,2)
            if coords.size(0) <= self.seq_len + 1:
                continue

            for i in range(self.seq_len, coords.size(0) - 1):
                past_abs   = coords[i - self.seq_len : i]   # (S,2)
                target_abs = coords[i]                      # (2,)

                lat0, lon0 = past_abs[-1]                   # window centre
                past_uv, _ = latlon_to_local_uv(
                    past_abs[:, 0], past_abs[:, 1], lat0, lon0,
                    half_side_mi=self.half_side_mi
                )
                target_uv, logJ = latlon_to_local_uv(
                    target_abs[0], target_abs[1], lat0, lon0,
                    half_side_mi=self.half_side_mi
                )

                centre_ll_norm = torch.tensor(
                    [lat0.item() / 90.0, lon0.item() / 180.0],
                    dtype=torch.float32,
                )

                yield past_uv, centre_ll_norm, target_uv, logJ
```

```python
# ─── Cell 3 : model definition ─────────────────────────────────────────────
class TransformerForecaster(nn.Module):
    def __init__(
        self,
        seq_len: int,
        d_model: int,
        nhead: int,
        num_layers: int,
        ff_hidden: int,
        fourier_m: int,
    ):
        super().__init__()
        # each timestep now has 4 features (u, v, lat_norm, lon_norm)
        self.input_proj = nn.Linear(4, d_model)
        self.pos_emb    = nn.Parameter(torch.randn(seq_len, d_model))
        enc_layer       = nn.TransformerEncoderLayer(d_model, nhead, ff_hidden, batch_first=True)
        self.encoder    = nn.TransformerEncoder(enc_layer, num_layers)
        self.fh         = FourierHead2D(
            dim_input=d_model, num_frequencies=fourier_m
        )

    def forward(self, past_uv, centre_ll, target_uv):
        """
        past_uv   : (B, S, 2)
        centre_ll : (B, 2)     (lat/90 , lon/180)
        target_uv : (B, 2)
        """
        B, S, _ = past_uv.shape
        centre_rep = centre_ll.unsqueeze(1).expand(-1, S, -1)      # (B,S,2)
        x = torch.cat([past_uv, centre_rep], dim=-1)               # (B,S,4)
    
        h = self.input_proj(x) + self.pos_emb.unsqueeze(0)         # (B,S,d)
        h = self.encoder(h)                                        # (B,S,d)  ← no transpose
        last = h[:, -1]                                            # (B,d)
        return self.fh(last, target_uv)
```

```python
# ─── Cell 4 : training loop with 10k-batch checkpoints ─────────────────────
from typing import Optional
import torch
import os

def train(
    df: pl.DataFrame,
    *,
    seq_len: int = 20,
    d_model: int = 128,
    nhead: int = 4,
    num_layers: int = 4,
    ff_hidden: int = 512,
    fourier_m: int = 64,
    batch_size: int = 64,
    lr: float = 1e-3,
    epochs: int = 5,
    device: Optional[str] = None,
    ckpt_dir: str = "checkpoints",
    model_dir = "random",
    ckpt_every: int = 10_000          # ← save every 10k batches
):
    device = device or ("cuda" if torch.cuda.is_available() else "cpu")
    print(f"🟡 Training on {device}", flush=True)

    # build dataloader
    ds = AISForecastIterableDataset(df, seq_len=seq_len)
    loader = DataLoader(
        ds, batch_size=batch_size, shuffle=False,
        drop_last=True, num_workers=0, pin_memory=True
    )

    # build model + optimiser
    model = TransformerForecaster(
        seq_len, d_model, nhead, num_layers,
        ff_hidden, fourier_m
    ).to(device)
    opt = optim.Adam(model.parameters(), lr=lr)

    # checkpoint folder
    os.makedirs(f"{ckpt_dir}/{model_dir}", exist_ok=True)
    global_step = 0

    for ep in range(1, epochs + 1):
        total_nll, count = 0.0, 0
        print(f"\n🟡 Epoch {ep}/{epochs}", flush=True)

        for b, (past_uv, centre_ll, target_uv, logJ) in enumerate(loader, 1):
            global_step += 1
            past_uv, centre_ll = past_uv.to(device), centre_ll.to(device)
            target_uv, logJ    = target_uv.to(device), logJ.to(device)

            pdf_uv = model(past_uv, centre_ll, target_uv)      # (B,)
            loss   = -(torch.log(pdf_uv + 1e-12) + logJ).mean()

            opt.zero_grad()
            loss.backward()
            opt.step()

            total_nll += loss.item() * past_uv.size(0)
            count     += past_uv.size(0)

            # --- checkpoint --------------------------------------------------
            if global_step % ckpt_every == 0:
                ckpt_path = f"{ckpt_dir}/{model_dir}/model_ep{ep:02d}_step{global_step:07d}.pt"
                torch.save(
                    {
                        "epoch": ep,
                        "step": global_step,
                        "model_state": model.state_dict(),
                        "optimizer_state": opt.state_dict(),
                    },
                    ckpt_path,
                )
                print(f"💾 Saved checkpoint → {ckpt_path}", flush=True)
            # ------------------------------------------------------------------

            if b % 1 == 0:
                print(f"    Batch {b:>5} | NLL {loss.item():.4f}", flush=True)

        print(f"🟢 Epoch {ep} | Avg NLL: {total_nll / count:.4f}", flush=True)

    return model
```

```python
# ─── Cell 5 : run everything ───────────────────────────────────────────────
df = load_cleaned_data("data/cleaned_partitioned_ais")

model = train(
    df,
    seq_len     = 20,
    d_model     = 128,
    nhead       = 4,
    num_layers  = 4,
    ff_hidden   = 512,
    fourier_m   = 128,     # far fewer freqs needed for a 100×100-mi window
    batch_size  = 64,
    lr          = 1e-5,
    epochs      = 100,
    model_dir = "seq=20-f=128"
)
print("✅ Training complete")
```

```python
# ╔══════════════════════════════════════════════════════════════════════════════╗
# ║   N_SAMPLES random zoomed forecasts – robust repeat-inside-plot version      ║
# ╚══════════════════════════════════════════════════════════════════════════════╝
import random, math
from pathlib import Path
import polars as pl
import torch
import matplotlib.pyplot as plt

# ───── ensure required globals exist (set sensible defaults) ─────────────
SEQ_LEN = globals().get("SEQ_LEN", 20)
HALF_SIDE_MI = globals().get("HALF_SIDE_MI", 50.0)
DEVICE = globals().get("DEVICE", torch.device("cuda" if torch.cuda.is_available() else "cpu"))
DATA_ROOT = globals().get("DATA_ROOT", Path.home() / "data" / "cleaned_partitioned_ais")
CKPT_PATH = globals().get("CKPT_PATH", Path.home() / "checkpoints" / "seq=20-f=32" / "model_ep01_step0020000.pt")

def latlon_to_uv(lat, lon, lat0, lon0, half_side_mi=HALF_SIDE_MI):
    R_mi = 69.0
    dx = R_mi * math.cos(math.radians(lat0)) * (lon - lon0)
    dy = R_mi * (lat - lat0)
    return dx / half_side_mi, dy / half_side_mi

# ───── demo hyper-params you can tweak ───────────────────────────────────
N_SAMPLES  = 20
ZOOM_FRAC  = 0.02           # ±2 % of 100-mi → ≈4×4-mi patch
GRID_ZOOM  = 200            # pixel resolution inside zoom patch
FOURIER_M  = 32             # must match checkpoint
D_MODEL    = 128            # keep in sync with training

# ───── build the SAME model & load checkpoint ────────────────────────────
from four_head_2D import FourierHead2D
class InferenceForecaster(torch.nn.Module):
    """Same encoder as training, but forward() omitted— we’ll
       compute the hidden vector manually inside the plot."""
    def __init__(self, seq_len, d_model, nhead, num_layers, ff_hidden,
                 fourier_m):
        super().__init__()
        self.seq_len   = seq_len
        self.input_proj= torch.nn.Linear(4, d_model)
        self.pos_emb   = torch.nn.Parameter(torch.randn(seq_len, d_model))
        enc = torch.nn.TransformerEncoderLayer(d_model, nhead, ff_hidden,
                                               batch_first=True)
        self.encoder   = torch.nn.TransformerEncoder(enc, num_layers)
        self.fh        = FourierHead2D(dim_input=d_model,
                                         num_frequencies=fourier_m)

model = InferenceForecaster(seq_len   = SEQ_LEN,
                            d_model   = D_MODEL,
                            nhead     = 4,
                            num_layers= 4,
                            ff_hidden = 512,
                            fourier_m = FOURIER_M).to(DEVICE)
state = torch.load(CKPT_PATH, map_location=DEVICE)
model.load_state_dict(state["model_state"])
model.eval()

# ───── load AIS tracks into memory ───────────────────────────────────────
parquet_files = list(Path(DATA_ROOT).rglob("*.parquet"))
df_all = pl.scan_parquet([str(f) for f in parquet_files]).collect()

tracks_by_vessel = {
    m: grp.sort("timestamp").select(["lat", "lon"]).to_numpy()
    for m, grp in df_all.group_by("mmsi")
    if grp.height > SEQ_LEN + 1
}
mmsis = list(tracks_by_vessel.keys())
print(f"🟢 vessels eligible: {len(mmsis)}")

# ───── pick random windows ───────────────────────────────────────────────
random.seed(123)
samples = []
while len(samples) < N_SAMPLES:
    m = random.choice(mmsis)
    tr = tracks_by_vessel[m]
    idx = random.randint(SEQ_LEN, tr.shape[0] - 1)
    samples.append((m, idx, tr))

# ───── helper to compute pdf grid (no forward() size mismatch) ───────────
def pdf_on_grid(past_uv, centre_ll, pts_uv):
    """Return PDF values for each (u,v) in pts_uv (N×2 tensor)."""
    with torch.no_grad():
        B, S, _ = past_uv.shape
        x = torch.cat([past_uv,
                       centre_ll.unsqueeze(1).expand(-1, S, -1)], dim=-1)
        h = model.input_proj(x) + model.pos_emb.unsqueeze(0)
        h = model.encoder(h)                         # (B,S,d)
        last = h[:, -1]                              # (1,d)
        pdf = model.fh(last.repeat(pts_uv.size(0), 1), pts_uv)
    return pdf

# ───── plot loop ─────────────────────────────────────────────────────────
for k, (mmsi, idx, track) in enumerate(samples, 1):
    past_abs   = torch.tensor(track[idx-SEQ_LEN:idx], dtype=torch.float32, device=DEVICE)
    target_abs = torch.tensor(track[idx],              dtype=torch.float32, device=DEVICE)
    lat0, lon0 = past_abs[-1]

    # local coords & centre vector
    past_uv = torch.tensor([latlon_to_uv(lat.item(), lon.item(),
                                         lat0.item(), lon0.item(), HALF_SIDE_MI)
                            for lat, lon in past_abs],
                           dtype=torch.float32, device=DEVICE).unsqueeze(0)
    centre_ll = torch.tensor([[lat0/90.0, lon0/180.0]],
                             dtype=torch.float32, device=DEVICE)

    # u,v grid for the zoom patch
    u = torch.linspace(-ZOOM_FRAC, ZOOM_FRAC, GRID_ZOOM, device=DEVICE)
    v = torch.linspace(-ZOOM_FRAC, ZOOM_FRAC, GRID_ZOOM, device=DEVICE)
    uu, vv = torch.meshgrid(u, v, indexing="xy")
    pts_uv = torch.stack([uu, vv], -1).view(-1, 2)

    # evaluate pdf and reshape
    pdf = pdf_on_grid(past_uv, centre_ll, pts_uv).view(GRID_ZOOM, GRID_ZOOM).cpu()

    # convert extent to lat/lon
    R_mi = 69.0
    lat_grid = lat0.item() + vv.cpu().numpy() * HALF_SIDE_MI / R_mi
    lon_grid = lon0.item() + uu.cpu().numpy() * HALF_SIDE_MI / \
               (R_mi * math.cos(math.radians(lat0.item())))
    extent = [lon_grid.min(), lon_grid.max(), lat_grid.min(), lat_grid.max()]

    # plot
    plt.figure(figsize=(5,4))
    plt.imshow(pdf, origin="lower", extent=extent, cmap="viridis", aspect="auto")
    plt.colorbar(label="PDF")
    plt.plot(past_abs[:,1].cpu(), past_abs[:,0].cpu(), 'w.-', label="Past track")
    plt.scatter(target_abs[1].cpu(), target_abs[0].cpu(),
                c="red", marker="x", label="True next")
    plt.title(f"Forecast {k}/{N_SAMPLES} – MMSI {mmsi}")
    plt.xlabel("Longitude"); plt.ylabel("Latitude")
    plt.legend()
    plt.tight_layout()
    plt.show()
```
